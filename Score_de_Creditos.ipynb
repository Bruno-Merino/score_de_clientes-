```python
# importando
import pandas as pd
import numpy
```

```python
# Passo 2: Visualizar a base de dados
tabela = pd.read_csv("/content/drive/MyDrive/Projeto de Analise de Crédito/clientes.csv")
display(tabela)
```

```python
# Resumo das Informações todas colunas da nossa tabela
print(tabela.info())
```

```python
# Preciso transformar  algumas colunas que não são numéricas em numericas que são
# ela: profissao mix_credito comportamento_pagamento e score_credito
# essa coluna vai ser a base vai ser a coluna base referencia não posso deletar ela
```

```python
# importando biblioteca de inteligencia artificial e fazendo as transformações
from sklearn.preprocessing import LabelEncoder
# Criar ocodificador pra transformar o tipo de dados
codificador = LabelEncoder()
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])
tabela["mix_credito"] = codificador.fit_transform(tabela["mix_credito"])
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"])
print(tabela.info())
```

```python
#Separar as colunas em dados de treino e dados de teste
# y nossa coluna referencia que não é numérica
y = tabela["score_credito"]

# axis = 1 -> são coluna
# axis = 2 -> são linha

# deletei a coluna score credito e id clientes
x = tabela.drop(["score_credito", "id_cliente"], axis=1)

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
#Criando um modelo de IA --> Score de Credito Ruin, Medio e Bom
# Arvore de Decisão
# KNN nearest neighbors


from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Escolha do melhor modelo mais acertivo calculando
# Arvore de decisão 82%
# KNN 73%
from sklearn.metrics import accuracy_score
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())
print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))
```
